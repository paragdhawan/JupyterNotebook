# TEMPLATE MATCHING


## Importing important libarires

```python
import imutils
import matplotlib.pyplot as plt
import cv2
import numpy as np
%matplotlib inline
```

## Reading Input Image

```python
img = cv2.imread('m1t.jpg')
print('Original Dimensions : ', img.shape)
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

## Converting Input Image to gray Scale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
```

## Reading Template

```python
template1 = cv2.imread('m2template.jpg', 0)
plt.imshow(template1)
```

## Denoising the image and template 

```python
img_gray = cv2.bilateralFilter(img_gray, 11, 17, 17)
template1 = cv2.bilateralFilter(template1, 11, 17, 17)
```

## Applying scaling and rotation with match template function ##

```python
# loop over the scales of the image
for scale in np.linspace(0.2, 1.0, 20)[::-1]:
        # resize the image according to the scale, and keep track
        # of the ratio of the resizing
    resized = imutils.resize(template1, width=int(template1.shape[1] * scale))
    for angle in np.arange(0, 360, 15):
        rotated = imutils.rotate(resized, angle)
        w, h = rotated.shape[::-1]

        # Perform match operations.
        res = cv2.matchTemplate(img_gray, rotated, cv2.TM_CCOEFF_NORMED)

        # Specify a threshold
        threshold = 0.8

        # Store the coordinates of matched area in a numpy array
        loc = np.where(res >= threshold)

        if loc:
            for pt in zip(*loc[::-1]):
                cv2.rectangle(
                    img2, pt, (pt[0] + w, pt[1] + h), (0, 255, 255), 2)

# cv2.imshow("image",img2)
# cv2.waitKey(0)
plt.imshow(img2)
```
